## 1. Write a function that returns the greater of two numbers

```python
# Your code here
def greater(a,b):
    """
    This function compares two numbers and returns the greater value.

    Example:
    greater(3, 8) 
    -> 8

    greater(5, 1)
    -> 5
    """
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    return a if a > b else b
```

```python
# Your code here
def greatest(lista):
    """
    This function accepts a list and return largest number in it.
    
    Example:
    greatest([1, 0, 5, 5, 7, 3]) -> 7
    greatest([2.3, 7, 8, 3, 1.1]) -> 8
    """
    if len(lista) == 0:
        raise ValueError("List cannot be empty")

    if not all(isinstance(num, (int, float)) for num in lista):
            raise TypeError("List must contains numbers")
        
    lista.sort()
    return lista[-1]
```

## 3. Write a function that sums all the elements on a list

```python
def sum_all(lst):
    """
    This function adds all the elements in a list and returns their sum.

    Example:
    sum_all([1, 4, 6]) -> 11
    sum_all([2, 6.9, 4]) -> 12.9
    """

    if not isinstance(lst, list):
        raise TypeError("Please input a list")

    if len(lst) == 0:
        raise ValueError("List cannot be empty")

    if not all(isinstance(num, (int, float)) for num in lst):
        raise TypeError("List must contain numbers")

    total_sum = 0
    for num in lst:
        total_sum += num

    return total_sum
```

```python
# This will test your function 
test_sum(sum_all)
```

## 4. Write another function that multiplies all the elements on a list

```python
def mult_all(lst):
    """
    This function multiply all the elements and return the total value.

    Example:

    List_1 = [1, 2, 3] -> 6
    List_2 = [1, 2.5, 2] -> 5
    """
    if not isinstance(lst,list):
        raise TypeError("Please provide a list as an argumnent")
    
    if len(lst) == 0:
        raise ValueError("Oops! You entered empty list. It should contain atleast one element.")
    
    if not all(isinstance(num, (int, float)) for num in lst):
        raise TypeError("List must contain only integers or floats.")
    
    mul_total = 1
    for num in lst:
        mul_total *= num
    return mul_total
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    fact = 1
    while n>1:
        fact *= n
        n-=1
    return fact
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
    freq = {}
    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    most_freq = max(freq.values())
    for key , value in freq.items():
        if value == most_freq:
            return key
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    freq_count = {}

    for char in string.replace(" ", "").lower():
        if "a" <= char <= "z":
            if char in freq_count:
                freq_count[char] += 1
            else:
                freq_count[char] = 1

    total_count = 0
    for key in freq_count.keys():
        total_count += 1

    if total_count == 26:
        return True
    else:
        return False
```

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):

    sort_list = []

    for char in string:
        if char == ",":
            sort_list.append(" ")
        else:
            sort_list.append(char)

    print(sort_list)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    if len(password) > 0:

        for char in password:
            if char.islower() > 1 and char.isupper() >= 1 and char.isdigt() > 1:
                return True
            else:
                return False

```
